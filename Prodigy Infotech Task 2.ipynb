```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips_data = sns.load_dataset("tips")
```

```python
print(tips_data.head())
```

```python
print(tips_data.isnull().sum())
```

```python
# Explore the distribution of numerical variables using histograms
plt.figure(figsize=(12, 8))
sns.histplot(data=tips_data, x='total_bill', bins=20, kde=True)
plt.xlabel('Total Bill Amount')
plt.ylabel('Frequency')
plt.title('Distribution of Total Bill Amounts')
plt.show()
```

```python
# Explore the relationship between total bill and tip amount
plt.figure(figsize=(12, 8))
sns.scatterplot(data=tips_data, x='total_bill', y='tip')
plt.xlabel('Total Bill Amount')
plt.ylabel('Tip Amount')
plt.title('Relationship between Total Bill and Tip Amount')
plt.show()
```

```python
# Explore the distribution of categorical variables using count plots
plt.figure(figsize=(12, 8))
sns.countplot(data=tips_data, x='day')
plt.xlabel('Day of the Week')
plt.ylabel('Count')
plt.title('Distribution of Days')
plt.show()
```

```python
# Explore the relationship between day and tip amount
plt.figure(figsize=(12, 8))
sns.boxplot(data=tips_data, x='day', y='tip')
plt.xlabel('Day of the Week')
plt.ylabel('Tip Amount')
plt.title('Relationship between Day and Tip Amount')
plt.show()
```
